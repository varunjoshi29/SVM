#### (e) XGBoost and Model Tree
#### In the case of a univariate tree, only one input dimension is used at a tree split. In a multivariate tree, or model tree, at a decision node all input dimensions can be used and thus it is more general. In univariate classification trees, majority polling is used at each node to determine the split of that node as the decision rule. In model trees, a (linear) model that relies on all of the variables is used to determine the split of that node (i.e. instead of using X j > s as the decision rule, one has P j Î² j X j > s . as the decision rule). Alternatively, in a regression tree, instead of using average in the region associated with each node, a linear regression model is used to determine the value associated with that node. One of the methods that can be used at each node is Logistic Regression. Because the number of variables is large in this problem, one can use L 1 -penalized logistic regression at each node. You can use XGBoost to fit the model tree. Determine Î± (the regularization term) using cross-validation. Train the model for the APS data set without compensation for class imbalance. **(f) Use SMOTE (Synthetic Minority Over-sampling Technique) to pre-process your data to compensate for class imbalance.  Train XGBosst with L1-penalized logistic regression at each node using the pre-processed data and repeat 1e. Do not forget that there is a right and a wrong way of cross validation here. Compare the uncompensated case with SMOTE case.**

**Right way is to use SMOTE inside cross-validation process**

```python
#https://machinelearningmastery.com/smote-oversampling-for-imbalanced-classification/
```

```python
smote_xgb_results = []
cross_val = KFold(n_splits=5, shuffle=True)
for alpha in alphas:
    xgb_tree = XGBClassifier(alpha=alpha, objective ='reg:logistic')
    steps = [('over', SMOTE()), ('model', xgb_tree)]
    pipeline = Pipeline(steps=steps)
    roc_auc_score = cross_val_score(pipeline, df_train_X, df_train_Y, cv=cross_val, scoring='roc_auc').mean()
    smote_xgb_results.append({"Alpha":alpha ,"roc_auc_score":roc_auc_score})
```

```python
pd.DataFrame(smote_xgb_results).sort_values(by=['roc_auc_score'], ascending=False)
```

### 2. ISLR 6.6.3

Ans:

#### (a)
(iv) Steadily decrease: As $s$ increases the constraint on $\beta$ 's decreases and the RSS reduces until we reach the least squares answer.

#### (b)
(ii) Decrease initially, and then eventually start increasing in a U shape: When $s = 0$, all  $\beta$ 's are $0$, the model is pretty simple and therefore has a very high test RSS. As we increase $s$, $\beta$ 's assume non-zero values and model starts to fit well on test data and therefore test RSS decreases. Eventually, as $\beta$ 's approach their full blown Ordinary Least Squares values, they become overfit to training data, thereby increasing test RSS. 

#### (c)
(iii) Steadily increase: When $s = 0$, we have become very strict on $\beta$ 's and the model effectively predicts a constant and has almost no variance. As we increase $s$, the models start to include more $\beta$ 's and their values start increasing. At this point, the values of $\beta$ s become highly dependent on training data, thus increasing the variance. 

#### (d)
(iv) Steadily decrease: When $s = 0$, the model effectively predicts a constant and hence the prediction is far from actual value. Thus bias is high. As $s$ increases, more $\beta$ 's become non-zero and thus the model continues to fit training data better. And thus, bias decreases.

#### (e)
(v) Remains constant: By assumption and definition, irreducible error is model independent and hence irrespective of the choice of $s$, hence remains constant.

```python
def plot_line(slope, intercept):
    axes = plt.gca()
    x_values = np.array(axes.get_xlim())
    y_values = intercept + slope * x_values
    plt.plot(x_values, y_values)
```

```python
def plot_dotted_line(slope, intercept):
    axes = plt.gca()
    x_values = np.array(axes.get_xlim())
    y_values = intercept + slope * x_values
    plt.plot(x_values, y_values,'--',c='purple')
```

```python
def add_point(scatter_plot, new_point, color):
    old_off = scatter_plot.get_offsets()
    new_off = np.concatenate([old_off,np.array(new_point, ndmin=2)])
    old_c = scatter_plot.get_facecolors()
    new_c = np.concatenate([old_c, np.array(matplotlib.colors.to_rgba(color), ndmin=2)])
    scatter_plot.set_offsets(new_off)
    scatter_plot.set_facecolors(new_c)
```
